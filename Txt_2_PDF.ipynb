```python
import os
import zipfile
from fpdf import FPDF

def unzip_file(zip_path, extract_to):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)

def txt_to_pdf(txt_file, pdf_file):
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Courier", size = 11)

    with open(txt_file, 'r') as file:
        for line in file:
            pdf.cell(200, 10, txt = line, ln = True)

    pdf.output(pdf_file)

def convert_txt_to_pdf_in_directory(directory):
    for foldername, subfolders, filenames in os.walk(directory):
        for filename in filenames:
            if filename.endswith('.txt'):
                txt_path = os.path.join(foldername, filename)
                pdf_path = os.path.splitext(txt_path)[0] + '.pdf'
                txt_to_pdf(txt_path, pdf_path)
                os.remove(txt_path)  # Delete the original text file

# Example usage
zip_path = '/content/SAM.gov Wage Determination.zip'
extract_to = '/content/pdfs'        # Replace with your extraction directory

unzip_file(zip_path, extract_to)
convert_txt_to_pdf_in_directory(extract_to)
```
